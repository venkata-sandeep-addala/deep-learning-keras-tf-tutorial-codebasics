```python
import numpy as np
```

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
def mse(y_true, y_predicted):
    total_error=0
    for yt, yp in zip(y_true, y_predicted):
        total_error+= (yt-yp)**2
    mse = total_error/len(y_predicted)
    return mse
```

```python
mse(y_true, y_predicted)
```

```python
np.mean(np.abs(y_true-y_predicted)**2)
```

```python
epsilon = 1e-15
epsilon
```

```python
y_predicted_new = [max(val, epsilon) for val in  y_predicted]
y_predicted_new
```

```python
y_predicted_new = [min(val, 1-epsilon) for val in  y_predicted_new]
y_predicted_new
```

```python
y_predicted_new = np.array(y_predicted_new)
```

```python
-np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```
